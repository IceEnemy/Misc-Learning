```python
import cv2
```

```python
# imread -> read image and store it in a variable
# imshow -> show the image in a window
# imwrite -> save an image in a given directory
image_path = "OtterGraduate.png"
image = cv2.imread(image_path)

#grayscale
gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

cv2.imshow('Hehe Sleepy Gray', gray_image)

cv2.imshow('Hehe Sleepy', image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
output_path = './processed_image.png'
cv2.imwrite(output_path, gray_image)
```
