```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([1,3,6,9,5,10])
y = np.array([2,4,5,7,10,12])

A = np.vstack([x,np.ones(len(x))]).T
y = y[:,np.newaxis]

pinv = np.linalg.pinv(A)
alpha = pinv.dot(y)
print(alpha)

plt.figure(figsize=(15,10))
plt.plot(x,y,"og",label="x y coords")
plt.plot(x,alpha[0]*x+alpha[1],"r",label="Least Square Regression")
plt.title("LSR")
plt.xlabel("x")
plt.ylabel("y")

plt.legend()
plt.show()
```
