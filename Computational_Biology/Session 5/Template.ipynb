## QUIZ PRACTICE (QUICK RECAP FROM SESSION 1 - SESSION 3)

### SESSION 1

```python
# Make new sequence from fasta.sequence
new_seq = seq_record.seq
print(new_seq)
```

```python
# Find length of the sequence Fasta File
length = len(new_seq)
print(length)
```

```python
# Make new DNA with 18 starting nucleotides from Fasta Sequence
new_dna = new_seq[:18]
print(new_dna)
```

```python
# Reverse the DNA
reverse_dna = new_dna[::-1]
print(reverse_dna)
```

```python
# Plot the DNA sequence nucleotide frequency
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np

# Count the frequency of each nucleotide

nucleotide_frequency = Counter(new_dna)
print(nucleotide_frequency)

# Define the colormap
colormap = plt.cm.autumn

# Get the colors for each nucleotide based on their order
colors = [colormap(i / float(len(nucleotide_frequency.keys()))) for i in range(len(nucleotide_frequency.keys()))]

# Plot the frequency of each nucleotide

plt.bar(nucleotide_frequency.keys(), nucleotide_frequency.values(), color = colors)
plt.show()
```

### SESSION 2

```python
# Find the GC & AT content of DNA
from Bio.SeqUtils import gc_fraction

gc_content = gc_fraction(new_dna) * 100
print(gc_content)

at_content = 100 - gc_content
print(at_content)
```

```python
# Find the melting temperature using Wallace, NN, GC of DNA
from Bio.SeqUtils import MeltingTemp as mt

melting_temp = mt.Tm_Wallace(new_dna)

print("Wallace: ", melting_temp)

#NN method
melting_temp = mt.Tm_NN(new_dna)

print("NN: ", melting_temp)

#GC method
melting_temp = mt.Tm_GC(new_dna)

print("GC: ", melting_temp)
```

```python
# Find the molecular weight of DNA
from Bio.SeqUtils import molecular_weight

molecular_weight = molecular_weight(new_dna)
print(molecular_weight)
```

### SESSION 3

```python
# Transcribe the DNA
from Bio.Seq import Seq

new_rna = new_dna.transcribe()

print(new_rna)
```

```python
# Translate the DNA
new_protein = new_dna.translate()
print(new_protein)
```

```python
# Convert Amino Acid to 3 letter codon
from Bio.SeqUtils import seq3

three_letter = seq3(new_protein)
print(three_letter)
```

```python
# Convert back from 3 letter codon to 1 letter codon

from Bio.SeqUtils import seq1

one_letter = seq1(three_letter)
print(one_letter)
```
