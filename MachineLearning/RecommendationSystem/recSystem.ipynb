```python
from sklearn.model_selection import train_test_split

user_music, unseen_music = train_test_split(dataset, test_size=0.5, random_state=42)

# user_music = dataset.sample(10)
user_music
```

```python
user_music_liked = user_music[user_music['liked'] == 1]
user_music_liked
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

user_preferences = scaler.fit_transform(user_music_liked.drop(columns=['liked'])).mean(axis=0)
user_preferences
```

```python
music_dict = unseen_music.drop(columns=['liked'])
music_dict = scaler.fit_transform(music_dict)
music_dict
```

```python
import numpy as np
from numpy import dot
from numpy.linalg import norm

def cosine_similarity(user_preferences, music):
    return dot(user_preferences, music) / (norm(user_preferences) * norm(music))
```

```python
cosine_similarity(user_preferences, music_dict[0])
```

```python
recc = pd.DataFrame(music_dict)
recc['similarity'] = recc.apply(lambda x: cosine_similarity(user_preferences, x), axis=1)
```
