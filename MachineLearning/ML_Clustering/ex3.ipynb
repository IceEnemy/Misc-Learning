```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('penguins.csv')

df.head()
```

```python
df.info()
```

```python
df = df.drop_duplicates()
df = df.dropna()

df.info()
```

```python
sns.pairplot(df)
```

```python
variables = ['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']

n_vars = len(variables)

fig, axes = plt.subplots(ncols=n_vars, figsize=(12,6))

for i,var in enumerate (variables):
    ax = axes[i]
    df.boxplot(column=var,ax=ax)
plt.show()
```

```python
def removeIQR(df,variables):
    for var in variables:
        Q1 = df[var].quantile(0.25)
        Q3 = df[var].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        df = df[(df[var] > lower_bound) & (df[var] < upper_bound)]
    return df

df = removeIQR(df, ['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g'])

variables = ['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']

n_vars = len(variables)

fig, axes = plt.subplots(ncols=n_vars, figsize=(12,6))

for i,var in enumerate (variables):
    ax = axes[i]
    df.boxplot(column=var,ax=ax)
plt.show()
```

```python
variables = ['sex']

for var in variables:
    counts = df[var].value_counts()
    counts.plot(kind='bar')
    plt.show()
```

```python
df = df[df['sex'] != '.']

df.sex.value_counts()
```

```python
from sklearn.preprocessing import MinMaxScaler

df[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']] = MinMaxScaler().fit_transform(df[['culmen_length_mm','culmen_depth_mm','flipper_length_mm','body_mass_g']])

df.describe()
```

```python
df = pd.get_dummies(df, columns=['sex'], dtype=int)

df.describe()
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

df = pd.DataFrame(pca.fit_transform(df),columns=['PC1','PC2'])

df.head()
```

```python
from sklearn.cluster import KMeans
WCSS = []

for i in range (1,11):
    kmeans = KMeans(n_clusters=i).fit(df)
    WCSS.append(kmeans.inertia_)

plt.figure(figsize=(12,6))
plt.plot(range(1,11), WCSS, marker='o')
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil = []

for i in range(2,11):
    kmeans = KMeans(n_clusters=i).fit(df)
    sil.append(silhouette_score(df,kmeans.labels_))

plt.figure(figsize=(12,6))
plt.plot(range(2,11), sil, marker='o')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4).fit(df)
df['Cluster'] = kmeans.labels_

plt.figure(figsize=(12,8))
plt.scatter(df['PC1'],df['PC2'],c=df['Cluster'])
plt.show()
```
