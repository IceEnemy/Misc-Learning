```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import sklearn as sk;
```

```python
dataset = pd.read_csv('iris.csv');
```

```python
dataset
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

dataset[['sepal.length', 'petal.width']] = scaler.fit_transform(dataset[['sepal.length', 'petal.width']])
# dataset['sepal.length'] = scaler.fit_transform(dataset['sepal.length'].values.reshape(-1, 1))
# dataset['petal.width'] = scaler.fit_transform(dataset['petal.width'].values.reshape(-1, 1))

varieties = dataset['variety'].unique()

colors = plt.cm.viridis(np.linspace(0, 1, len(varieties)))

color_map = dict(zip(varieties, colors))

for variety in varieties:
    subset = dataset[dataset['variety'] == variety]
    plt.scatter(subset['sepal.length'], subset['petal.width'], color=color_map[variety], label=variety)

plt.legend(title = 'Variety')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.title('Correlation of Sepal Length and Petal Width with Variety')

plt.show()



```
