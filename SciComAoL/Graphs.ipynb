```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y_odd = np.array([1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999])
x_odd = np.array([14.1999, 14.2411, 14.0342, 14.2696, 14.197, 14.3055, 14.1853, 14.3577, 14.4187, 14.3438])
y_even = np.array([1982, 1984, 1986, 1988, 1990, 1992, 1994, 1996, 1998])
x_even = np.array([14.2947, 14.1255, 14.0535, 14.1851, 14.2542, 14.2996, 14.2839, 14.4037, 14.3998])

year1 = np.array([1981,1982,1983,1984,1985,1986,1987])
year2 = np.array([1987,1988,1989,1990,1991,1992,1993])
year3 = np.array([1993,1994,1995,1996,1997,1998,1999])

plt.figure(figsize=(20,10))

plt.plot(y_odd,x_odd,"og",label = "Odd Years")
plt.plot(y_even,x_even,"ob",label = "Even Years")
plt.plot(year1,14.1999 + 0.0206*(year1-1981) - 0.031*(year1-1981)*(year1-1983) + 0.0144*(year1-1981)*(year1-1983)*(year1-1985),"r",label="Interpolation")
plt.plot(year2,14.2696 - 0.0363*(year2-1987) + 0.0226*(year2-1987)*(year2-1989) - 0.0085*(year2-1987)*(year2-1989)*(year2-1991),"r")
plt.plot(year3,14.1853 + 0.0862*(year3-1993) - 0.0139*(year3-1993)*(year3-1995) - 0.0005*(year3-1993)*(year3-1995)*(year3-1997),"r")
plt.plot()

plt.title("Cubic Interpolation")
plt.xlabel("Years (y)")
plt.ylabel("Temperature (x)")

plt.xticks(np.arange(1981, 2000, 1))
plt.grid(True)
plt.legend()
plt.show()
```
